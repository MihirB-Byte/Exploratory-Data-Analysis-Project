```python
### Start Time

df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde = False, norm_hist = True)
```

```python
# - High percentage of accidents occur between 3 PM to 5 PM (Probably people in hurry to get out of work)

# - Next highest percentage is between 7 AM and 9 AM( People in hurry to get to work)
```

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde = False, norm_hist = True)
```

```python
# - Is the distribution of accidents by hour the same on weekends as on weekdays?
```

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek ==6]
sns.distplot(sundays_start_time.dt.hour, bins=24, kde = False, norm_hist = True)
```

```python
mondays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek ==0]
sns.distplot(mondays_start_time.dt.hour, bins=24, kde = False, norm_hist = True)
```

```python
# On Sundays the peak occures between 10 AM and 3 PM, unlike the weekdays.
```

```python
sns.distplot(mondays_start_time.dt.month, bins=12, kde = False, norm_hist = True)
```

```python
sns.distplot(df.Start_Time.dt.month, bins=12, kde = False, norm_hist = True)
```

```python
df_2019 = df[df.Start_Time.dt.year == 2020]

sns.distplot(df_2019.Start_Time.dt.month, bins=12, kde = False, norm_hist = True)
```
